```python
athlete_data['Age'] = athlete_data['Age'].fillna(athlete_data['Age'].mean())
athlete_data['Height'] = athlete_data['Height'].fillna(athlete_data['Height'].mean())
athlete_data['Weight'] = athlete_data['Weight'].fillna(athlete_data['Weight'].mean())
```

Popular Sports based on Athlete participation

```python
athlete_data['Sport'].value_counts().head(10)
```

```python
top_sports = athlete_data['Sport'].value_counts().head(10)
plt.figure(figsize=(8,4))
sns.barplot(y =top_sports.index, x = top_sports.values, hue=top_sports.index, palette='husl', dodge=False)
plt.title("Popularity of top 10 Sports")
plt.ylabel("Sport")
plt.xlabel("Number of Athlete")
plt.show()
```

```python
medal_count = athlete_data['NOC'].value_counts().head(10)
plt.figure(figsize=(8,4))
sns.barplot(x = medal_count.index, y = medal_count.values, color='gold')
plt.title('Top 10 Countries by Total Medals')
plt.xlabel('Country (NOC)')
plt.ylabel('Number of Medals')
plt.show()
```

India's Medal count over the years.


```python
athlete_data.head(1)
```

```python
Medal_country = athlete_data[athlete_data['Team'] == 'India']
```

```python
india_medal = Medal_country.groupby('Year')['Medal'].count()
```

```python
india_medal.index
```

```python
india_medal.values
```

```python
sns.lineplot(x = india_medal.index, y = india_medal.values)
plt.title("India's Medal Count Over the Years")
plt.xlabel("Year")
plt.ylabel("No. of Medals")
plt.show()
```
